```python
fig,ax=plt.subplots(1,2,layout='constrained')
ax[0].scatter('ttl_jobs','avg_savings',data=nta)
ax[1].set_yscale('log')
ax[1].scatter('ttl_jobs','avg_savings',data=nta)
```

```python
#histogram of the log of the average total savings.
avg=nta['avg_savings']
"""
plt.rcParams['figure.autolayout']=True
plt.hist(np.log(avg),density=True,log=True)
plt.show()
"""
#nta['avg_savings'].plot(kind='hist',logx=True)
#nta['avg_savings'].hist(bins=50,density=True)
#plt.show()
fig,ax=plt.subplots()
n,bins,patches=ax.hist(np.log(avg),50,density=True,facecolor='C0',alpha=0.75)
ax.grid(True)
ax.set_xlabel('The log of the average total savings')
ax.set_ylabel('Probability')
```
